```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Indirect(y, t, TF1, k1, k2, k3, k4, k5, k6, Kd1, Kd2, Kd11, total_time):
    if t > total_time / 3:
        TF1 = 0
    if t > total_time * 2/3:
        TF1 = 1
    d_dt = np.zeros(3)
    d_dt[0] = (1-(1/(1+(y[2]/Kd2))))*k1 - k2*y[0]
    d_dt[1] = (1-(1/(1+(TF1/Kd11))))*k3 - k4*y[1]
    d_dt[2] = k5*y[1] - k6*y[2]
    return d_dt
```

```python
def integrate_Indirect(y0, time_steps, params):
    total_time = time_steps[-1]
    result = odeint(
        lambda y, t: Indirect(y, t, params['TF1'], params['k1'], params['k2'], params['k3'], params['k4'],
                             params['k5'], params['k6'], params['Kd1'], params['Kd2'], params['Kd11'], total_time),
        y0, time_steps
    )
    return result
```

```python
params = {'TF1':1, 'k1':10, 'k2':.1, 'k3':10, 'k4':.1, 'k5':10, 'k6':.1, 'Kd1':10, 'Kd2':10, 'Kd11':10}
y0 = [0, 0, 0]
time_steps = np.linspace(0, 1000, 1000)
```

```python
data = integrate_Indirect(y0, time_steps, params)
```

```python
plt.plot(time_steps, data[:,0], label = '[mRNA]')
plt.axvline(x = 1000/3, color = 'r', label = 'TF1 off')
plt.axvline(x = 1000 * 2/3, color = 'g', label = 'TF1 on')
# plt.plot(time_steps, data[: ,1], label = '[TF]')
plt.legend(loc = 'best')
```
