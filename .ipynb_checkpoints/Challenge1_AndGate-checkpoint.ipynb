```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AndGate(y, t, TF1, k1, k2, k3, k4, k5, k6, Kd1, Kd2, Kd11):
    d_dt = np.zeros(3)
    d_dt[0] = (1 - (1 / (1 + (TF1 / Kd1)))) * (1 - (1 / (1 + (y[2] / Kd2)))) * k1 - k2 * y[0]
    d_dt[1] = (1 - (1 / (1 + (TF1 / Kd11)))) * k3 - k4 * y[1]
    d_dt[2] = k5 * y[1] - k6 * y[2]
    return d_dt
```

```python
TF1 = 1
k1 = 1e-2
k2 = 1e-2
k3 = 1e-2
k4 = 1e-2
k5 = 1e-2
k6 = 1e-2
Kd1 = 1e-3
Kd2 = 1e-3
Kd11 = 1e-3
y0 = [0, 0, 0]
time_steps = np.linspace(0, 1000, 1000)
```

```python
data = odeint(AndGate, y0, time_steps, args=(TF1, k1, k2, k3, k4, k5, k6, Kd1, Kd2, Kd11)) 
```

```python
plt.plot(time_steps, data[:,0], label = '[mRNA]')
# plt.plot(time_steps, data[: ,1], label = '[TF]')
plt.legend(loc = 'best')
```
